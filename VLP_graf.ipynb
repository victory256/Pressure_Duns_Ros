```python
from Pressure_Duns_Ros import *
import matplotlib.pyplot as plt

shag = 100
q_liq_min = 190
q_liq_max = 2000
q_liq = range(q_liq_min, q_liq_max, shag)
R_s_input = 500
WC_input = 0.7
L_input = 2000
delta_P = []
for i in q_liq:
    delta_P.append(calc_pressure(i, R_s_input, WC_input,L_input))

P_whp = 0
P_bhp = 100
if P_whp == 0:
    print("при забойном P_bhp =", P_bhp,"атм и Q_ж =",q_liq_max,"ст.м^3/сут -> устьевое давление P_whp =", P_bhp - calc_pressure(q_liq_max, R_s_input, WC_input,L_input),"атм")
if P_bhp == 0:
    print("при устьевом P_whp =", P_whp,"атм и Q_ж =",q_liq_max,"ст.м^3/сут -> забойное давление P_bhp =", P_whp + calc_pressure(q_liq_max, R_s_input, WC_input,L_input),"атм")

L_mass = range(100, 2000, shag)
press_mass = [calc_pressure(q_liq_max, R_s_input, WC_input,L_iter) for L_iter in L_mass]
fig = plt.figure()

plt.subplot(1, 2, 1)
plt.title("delta P для Q_liq = 2000 ст.м^3/сут")
plt.xlabel("L, м")
plt.ylabel("delta Р, бар")
plt.plot(L_mass, press_mass)

plt.subplot(1, 2, 2)
plt.title("VLP")
plt.xlabel("q_ж, ст.м^3/сут")
plt.ylabel("delta Р, бар")
plt.plot(q_liq, delta_P)
```
